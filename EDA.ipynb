```python
plt.figure(figsize=(14, 8))  
corr_matrix = df[numeric_cols].corr()

# Create a mask to hide upper triangle (optional for readability)
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Drawing the heatmap with better aesthetics
sns.heatmap(corr_matrix, 
            annot=True, 
            cmap="coolwarm", 
            fmt=".2f", 
            linewidths=0.5, 
            mask=mask,  # Hide upper triangle
            annot_kws={"size": 10})  

plt.title("Correlation Heatmap", fontsize=14)
plt.show()
```

#### Taxes are directly proportional to total sales... as there is The strong correlation (close to 1) between Total Sales Tax @ 18% and Total Amount Excluding Taxes....confirms that taxes are consistently applied as a percentage of sales.

#### The correlations between Quantity_Product_1 and Quantity_Product_2, etc., are moderate (0.2 - 0.5).
#### Reason: Customers might not be consistently purchasing multiple products together.



```python
# Top 10 Customer Businesses
df["Customer_Business"].value_counts().head(10).plot(kind="barh", figsize=(10, 5), color="royalblue")
plt.xlabel("Count")
plt.ylabel("Business Type")
plt.title("Top 10 Business Types")
plt.show()
```

```python
### Distribution of Transaction Amounts (Total Amount Including Taxes)
plt.figure(figsize=(10, 5))
sns.histplot(df["Total Amount Including Taxes"], bins=30, kde=True, color="royalblue")
plt.xlabel("Total Amount (Including Taxes)")
plt.ylabel("Frequency")
plt.title("Distribution of Total Transaction Amounts")
plt.show()
```

#### A right-skewed distribution suggests that a few high-value transactions contribute significantly.

```python
#### Box plot of total amount including taxes
plt.figure(figsize=(8, 4))
sns.boxplot(x=df["Total Amount Including Taxes"], color="coral")
plt.xlabel("Total Amount (Including Taxes)")
plt.title("Box Plot of Transaction Amounts")
plt.show()
```

##### transaction amounts have a positively skewed distribution.
##### here outliers are the high-value transactions that significantly impact total revenue.

```python
### Most frequently purchased products (Identify top selling products which can help in stock management)
product_counts = df[["Product 1", "Product 2", "Product 3", "Product 4"]].melt().value_counts().head(10)
product_counts.plot(kind="bar", figsize=(10, 5), color="green")
plt.xlabel("Product")
plt.ylabel("Count")
plt.title("Top 10 Most Purchased Products")
plt.show()
```

```python
### Sales Trend Over Time
df.set_index("Date")["Total Amount Including Taxes"].resample("M").sum().plot(figsize=(12, 6), marker="o", color="purple")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.title("Monthly Sales Trend")
plt.show()
```

##### Consistent Sales with Fluctuations (2018–2023)  
##### There are some peaks around 2020 and 2021, possibly due to seasonal demand or promotions.

```python
##### Check if sales are driven by repeat customers or one-time buyers.
repeat_customers = df["Customer_NAME"].value_counts()
plt.figure(figsize=(8, 4))
sns.histplot(repeat_customers, bins=20, kde=True, color="blue")
plt.xlabel("Number of Transactions per Customer")
plt.ylabel("Number of Customers")
plt.title("Customer Purchase Frequency Distribution")
plt.show()
```

##### The business relies on a large number of one-time buyers rather than repeat customers.

```python
# 2. Total Amount Excluding Taxes vs. Including Taxes
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Total Amount Excluding Taxes"], y=df["Total Amount Including Taxes"], color="darkorange", alpha=0.6)
plt.xlabel("Total Amount (Excluding Taxes)")
plt.ylabel("Total Amount (Including Taxes)")
plt.title("Excluding vs. Including Taxes")
plt.show()
```

##### show a perfect linear relationship, meaning that the tax rate is consistently applied across all transactions.

```python
# Create Total_Quantity column
df_copy=df
df_copy["Total_Quantity"] = df[
    ["Quantity_Product_1", "Quantity_Product_2", "Quantity_Product_3", "Quantity_Product_4"]
].sum(axis=1)

# Plot Total Quantity vs. Total Amount Including Taxes
plt.figure(figsize=(7, 5))
sns.scatterplot(x=df_copy["Total_Quantity"], y=df_copy["Total Amount Including Taxes"], color="blue", alpha=0.6)

plt.xlabel("Total Quantity Sold")
plt.ylabel("Total Amount (Including Taxes)")
plt.title("Total Quantity vs. Total Amount (Including Taxes)")
plt.show()
```

#### As the total quantity sold increases, the total amount (including taxes) also increases.

```python
#### Top Customers by Purchase Amount
# Group by Customer and sum the Total Amount (Including Taxes)
top_customers = df.groupby("Customer_NAME")["Total Amount Including Taxes"].sum().nlargest(10)

# Plot
plt.figure(figsize=(10, 5))
top_customers.sort_values().plot(kind="barh", color="green")

# Labels and title
plt.xlabel("Total Purchase Amount (Including Taxes)")
plt.ylabel("Customer Name")
plt.title("Top 10 Customers by Total Purchase Amount")
plt.show()
```
