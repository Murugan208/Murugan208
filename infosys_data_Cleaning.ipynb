```python
# Check missing values
missing_values = df.isnull().sum()
print("\nMissing Values:\n", missing_values[missing_values > 0])
```

```python
### Fill Missing Values for NAME and Customer_NAME with "Unknown"
df['NAME'].fillna("Unknown", inplace=True)
df['Customer_NAME'].fillna("Unknown", inplace=True)
```

```python
# Check missing values
missing_values = df.isnull().sum()
print("\nMissing Values:\n", missing_values[missing_values > 0])
```

```python
missing_percentage = (df.isnull().sum() / len(df)) * 100
print(missing_percentage)
```

```python
## imputing product 1,2,3,4 with mode since they are categorical
df['Product 1'].fillna(df['Product 1'].mode()[0], inplace=True)
df['Product 2'].fillna(df['Product 1'].mode()[0], inplace=True)
df['Product 3'].fillna(df['Product 1'].mode()[0], inplace=True)
df['Product 4'].fillna(df['Product 1'].mode()[0], inplace=True)
```

```python
# Check missing values
missing_values = df.isnull().sum()
print("\nMissing Values:\n", missing_values[missing_values > 0])
```
